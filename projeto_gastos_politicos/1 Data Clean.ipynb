```python
import pandas as pd
```

```python
df = pd.read_csv('ceaps_dataset.csv',skiprows=1, delimiter=';', encoding='latin1')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.dtypes
```

```python
df['DATA'] = df['DATA'].str.replace('/', '-')
```

```python
df['DATA'] = df['DATA'].str.replace('5017','2017')
df['DATA'] = df['DATA'].str.replace('3016','2016')
df['DATA'] = df['DATA'].str.replace('206','2016')
df['DATA'] = df['DATA'].str.replace('216','2016')
df['DATA'] = df['DATA'].str.replace('200','2016')
```

```python
#pd.to_datetime(df['DATA'], errors='coerce').isnull().sum()
df['DATA'] = pd.to_datetime(df['DATA'])
```

```python
df['DIA'] = df['DATA'].dt.day
```

```python
df.head()
```

```python
df['VALOR_REEMBOLSADO'] = df['VALOR_REEMBOLSADO'].str.replace(',','.')
```

```python
df['VALOR_REEMBOLSADO'] = pd.to_numeric(df['VALOR_REEMBOLSADO'])
```

```python
df['MAIOR_QUE_100'] = 0
```

```python
df.loc[df['VALOR_REEMBOLSADO'] > 100,'MAIOR_QUE_100'] = 1
```

```python
df.head()
```

```python
df.columns
```

```python
df = df.filter(['ANO', 'MES','DIA', 'SENADOR', 'TIPO_DESPESA', 'CNPJ_CPF', 'FORNECEDOR',
       'DOCUMENTO', 'DATA', 'DETALHAMENTO', 'VALOR_REEMBOLSADO',
       'MAIOR_QUE_100'])
```

```python
df['TIPO_DESPESA'].unique()
```

```python
mapper = {
          'Aluguel de imóveis para escritório político, compreendendo despesas concernentes a eles.':'Aluguel de Imoveis',
          'Aquisicao de material de consumo para uso no escritório político, inclusive aquisição ou locação de software, despesas postais, aquisição de publicações, locação de móveis e de equipamentos. ':'Aquisicao de material',
          'Contratação de consultorias, assessorias, pesquisas, trabalhos técnicos e outros serviços de apoio ao exercício do mandato parlamentar':'Contratacoes',
          'Locomoção, hospedagem, alimentação, combustíveis e lubrificantes': 'Logistica',
          'Passagens aéreas, aquáticas e terrestres nacionais': 'Passagens',
          'Divulgação da atividade parlamentar':'Divulgacao',
          'Serviços de Segurança Privada':"Seguranca"}
df['TIPO_DESPESA'] = df['TIPO_DESPESA'].map(mapper)
```
